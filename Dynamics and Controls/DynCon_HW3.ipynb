```python
# Constants
x_F1 = sp.Matrix([[1],[0],[0]])
y_F1 = sp.Matrix([[0],[1],[0]])
z_F1 = sp.Matrix([[0],[0],[1]])

theta_x, theta_y, theta_z = sp.symbols('theta_x theta_y theta_z', real=True)
theta_x_dot, theta_y_dot, theta_z_dot = sp.symbols('theta_x_dot theta_y_dot theta_z_dot', real=True)

C_x = sp.Matrix([[1, 0, 0],
                [0, sp.cos(theta_x), sp.sin(theta_x)],
                [0, -sp.sin(theta_x), sp.cos(theta_x)]]) #type: ignore

C_y = sp.Matrix([[sp.cos(theta_y), 0, -sp.sin(theta_y)], #type: ignore
                [0, 1, 0],
                [sp.sin(theta_y), 0, sp.cos(theta_y)]])

C_z = sp.Matrix([[sp.cos(theta_z), sp.sin(theta_z), 0],
                [-sp.sin(theta_z), sp.cos(theta_z), 0], #type: ignore
                [0, 0, 1]])

C_21 = C_x @ C_z @ C_y

y_F1_inF2 = C_21 @ y_F1
z_F1_inF2 = C_x @ C_z @ z_F1
x_F1_inF2 = C_x @ x_F1

B = sp.Matrix.hstack(x_F1_inF2, y_F1_inF2, z_F1_inF2)
dotmatrix = sp.Matrix([[theta_x_dot], [theta_y_dot], [theta_z_dot]])
w_21 = B @ dotmatrix

B_smallangle = sp.Matrix([[1, theta_z, 0], [0, 1, theta_x],[0, -theta_x, 1]])

display(Math(r"\mathbf{\omega}_{21} = " + latex(B) + latex(dotmatrix)))
print('Now using small angle approximation')
display(Math(r"\mathbf{\omega}_{21} = " + latex(B_smallangle) + latex(dotmatrix)))
```
